```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "ibm.csv"  # Update the path if needed
df = pd.read_csv('C:/Users/Hp/Desktop/New folder/unified mentors/IBM HR Analytics/ibm.csv')
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df, palette="coolwarm")
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color="royalblue")
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='MonthlyIncome', y='JobRole', data=df, palette="viridis")
plt.title("Monthly Income by Job Role")
plt.xlabel("Monthly Income")
plt.ylabel("Job Role")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="JobSatisfaction", y=df["Attrition"].apply(lambda x: 1 if x == "Yes" else 0), data=df, ci=None, palette="magma")
plt.title("Job Satisfaction vs Attrition Rate")
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Attrition Rate")
plt.show()
```
