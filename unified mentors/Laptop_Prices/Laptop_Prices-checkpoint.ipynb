```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "laptop_prices.csv"
df = pd.read_csv("C:/Users/Hp/Desktop/unified mentors/Laptop_Prices/laptop_prices-checkpoint.csv")
```

```python
# Convert categorical columns
binary_columns = ['Touchscreen', 'IPSpanel', 'RetinaDisplay']
for col in binary_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
# 1. Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Price_euros'], bins=30, kde=True, color='blue')
plt.title('Laptop Price Distribution')
plt.xlabel('Price (Euros)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Top 10 Laptop Brands by Count
plt.figure(figsize=(12, 5))
df['Company'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Laptop Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Laptop Types Distribution
plt.figure(figsize=(8, 8))
df['TypeName'].value_counts().plot.pie(autopct='%1.1f%%', cmap='coolwarm')
plt.title('Laptop Type Distribution')
plt.ylabel('')
plt.show()
```

```python
# 4. Relationship Between RAM and Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Ram'], y=df['Price_euros'], hue=df['Company'], alpha=0.7)
plt.title('RAM vs Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# 5. Effect of Screen Size on Price
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Inches'], y=df['Price_euros'], marker='o', color='red')
plt.title('Screen Size vs Price')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# 6. CPU Frequency Across Brands
plt.figure(figsize=(12, 5))
sns.boxplot(x=df['CPU_company'], y=df['CPU_freq'], palette='Set2')
plt.title('CPU Frequency by Brand')
plt.xlabel('CPU Brand')
plt.ylabel('CPU Frequency (GHz)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Primary Storage Type vs Price
plt.figure(figsize=(10, 5))
sns.violinplot(x=df['PrimaryStorageType'], y=df['Price_euros'], palette='muted')
plt.title('Primary Storage Type vs Price')
plt.xlabel('Primary Storage Type')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# 8. Weight vs Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Weight'], y=df['Price_euros'], alpha=0.7, color='purple')
plt.title('Weight vs Price')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
# 9. Operating System Distribution
plt.figure(figsize=(10, 5))
df['OS'].value_counts().plot(kind='bar', color='orange')
plt.title('Operating System Distribution')
plt.xlabel('OS')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. GPU Brand vs Price
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['GPU_company'], y=df['Price_euros'], palette='coolwarm')
plt.title('GPU Brand vs Price')
plt.xlabel('GPU Brand')
plt.ylabel('Price (Euros)')
plt.xticks(rotation=45)
plt.show()
```
