```python
# Load dataset
file_path = "climate_nasa.csv"  # Update this with your actual file path
df = pd.read_csv("C:/Users/Hp/Desktop/unified mentors/Climate_Change/climate_nasa (1).csv")
```

```python
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# Handle missing values
df['commentsCount'].fillna(0, inplace=True)
df['text'].fillna("", inplace=True)
```

```python
# Extract useful date features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
```

```python
# Plot distribution of likes and comments
plt.figure(figsize=(12, 5))
sns.histplot(df['likesCount'], bins=30, kde=True, color='blue')
plt.title('Distribution of Likes Count')
plt.xlabel('Likes Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df['commentsCount'], bins=30, kde=True, color='red')
plt.title('Distribution of Comments Count')
plt.xlabel('Comments Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Time series plot of likes and comments
df_time = df.groupby('date').sum()
plt.figure(figsize=(15, 6))
plt.plot(df_time.index, df_time['likesCount'], label='Likes', color='blue')
plt.plot(df_time.index, df_time['commentsCount'], label='Comments', color='red')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Likes and Comments Over Time')
plt.legend()
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['likesCount'], color='blue')
plt.title('Boxplot of Likes Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['commentsCount'], color='red')
plt.title('Boxplot of Comments Count')
plt.show()
```

```python
# Generate a word cloud for post text
text_data = " ".join(df['text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud of Post Text')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df[['likesCount', 'commentsCount']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of likes vs comments
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['likesCount'], y=df['commentsCount'], color='green', alpha=0.5)
plt.title('Scatter Plot of Likes vs Comments')
plt.xlabel('Likes Count')
plt.ylabel('Comments Count')
plt.show()
```

```python
# Monthly trend of likes and comments
df_monthly = df.groupby(['year', 'month'])[['likesCount', 'commentsCount']].sum().reset_index()
plt.figure(figsize=(12, 5))
sns.lineplot(x=df_monthly.index, y=df_monthly['likesCount'], label='Likes', color='blue')
sns.lineplot(x=df_monthly.index, y=df_monthly['commentsCount'], label='Comments', color='red')
plt.title('Monthly Trend of Likes and Comments')
plt.xlabel('Time (Months)')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Hourly engagement pattern
plt.figure(figsize=(12, 5))
sns.barplot(x=df['hour'], y=df['likesCount'], color='blue')
plt.title('Hourly Engagement Pattern (Likes)')
plt.xlabel('Hour of the Day')
plt.ylabel('Likes Count')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=df['hour'], y=df['commentsCount'], color='red')
plt.title('Hourly Engagement Pattern (Comments)')
plt.xlabel('Hour of the Day')
plt.ylabel('Comments Count')
plt.show()
```
