```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/Sunil Kannur/OneDrive/Desktop/googleplaystore.csv")
```

```python
# Data Cleaning
# Handle missing values in Rating
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
```

```python
# Convert Reviews to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
# Clean Size column
df['Size'] = df['Size'].str.replace('M', '')
df['Size'] = df['Size'].str.replace('k', 'e-3')  # Convert KB to MB
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
```

```python
# Clean Installs column
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
# Clean Price column
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
# Set style for plots
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
```

```python
# 1. Top Categories by Number of Apps
plt.subplot(2,2,1)
top_categories = df['Category'].value_counts().nlargest(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 App Categories')
plt.xlabel('Number of Apps')
```

```python
# 2. Rating Distribution
plt.subplot(2,2,2)
sns.histplot(df['Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
```

```python
# 3. Free vs Paid Apps
plt.subplot(2,2,3)
type_counts = df['Type'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Proportion of Free vs Paid Apps')
```

```python
# 4. Top Genres by Installs
plt.subplot(2,2,4)
top_genres = df.groupby('Genres')['Installs'].sum().nlargest(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='rocket')
plt.title('Top 10 Genres by Installs')
plt.xlabel('Total Installs (Billions)')
plt.ticklabel_format(style='plain', axis='x')
```

```python
plt.tight_layout()
plt.show()
```

```python
# 5. Relationship between Size and Rating
plt.figure(figsize=(10,6))
sns.scatterplot(x='Size', y='Rating', data=df, alpha=0.6, color='purple')
plt.title('App Size vs Rating')
plt.xlabel('Size (MB)')
plt.ylabel('Rating')
plt.show()
```

```python
# 6. Price Distribution in Paid Apps
paid_apps = df[df['Type'] == 'Paid']
plt.figure(figsize=(10,6))
sns.histplot(paid_apps['Price'], bins=30, color='orange')
plt.title('Distribution of Prices for Paid Apps')
plt.xlabel('Price ($)')
plt.xlim(0, 50)  # Limit to $50 for better visibility
plt.show()
```

```python
# 7. Content Rating Distribution
plt.figure(figsize=(10,6))
content_ratings = df['Content Rating'].value_counts()
sns.barplot(x=content_ratings.values, y=content_ratings.index, palette='coolwarm')
plt.title('Content Rating Distribution')
plt.xlabel('Number of Apps')
plt.show()
```
